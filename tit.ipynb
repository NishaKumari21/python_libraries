```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
print("Importing all module")
```

Taking Data from the Titanic

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

This is Line plot of the Titanic data--->
Customizing
Here i used the marker as diamond and color as pink

# Line Plot

```python

plt.plot(titanic['who'], titanic['pclass'],marker="D",ms=10,color="pink")
plt.xlabel("Who")
plt.ylabel("Pclass")
plt.show()
```

# Bar Plot

```python
sns.barplot(x='who', y='fare', data=titanic,color="orange")
plt.xlabel("Who")
plt.ylabel("fare")
plt.show()
```

```python
sns.pieplot(x='who', y='pclass', data=titanic)
plt.xlabel("Who")
plt.ylabel("Pclass")
plt.show()
```

 # Scatter plot

```python

plt.scatter(x='fare',y='age',marker="o",color='pink')

plt.title("Scatter plot")
plt.show()
```

 # Histogram

```python
sns.histplot(x='age', data=titanic,color="pink",ec="blue")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

 # Box plot

```python
sns.boxplot(x='who', y='age', data=titanic,color="yellow")
plt.xlabel("Who")
plt.ylabel("Age")
plt.show()
```

This is relational plot of Kind:Scatter plot

```python
sns.relplot(x='fare',y='sex',data=titanic,color="yellow")
plt.show()
```

```python
sns.relplot(x='age',y='sex',data=titanic,color="yellow",hue='alive')
plt.show()
```

```python
sns.relplot(x='age',y='sex',data=titanic,color="yellow",hue='alive',style='alone')
plt.show()
```

```python
sns.relplot(x='age',y='sex',data=titanic,color="yellow",hue='alive',col='embarked')
plt.show()
```

This is relational plot of Kind:Line plot

```python
sns.relplot(x='who',y='age',data=titanic,hue='sex',kind='line')
plt.show()
```

sns.catplot() function in seaborn is used to create categorical plots. It provides a flexible interface for creating various types of plots, such as 
 stripplots,


  swarmplots,


  boxplots, 


  violinplots,


   bar plots, 


    point plots.

This is Categorical plot of kind-'strip'

```python
sns.catplot(x='age',y='who',data=titanic,hue='sex')
plt.show()
```

This is Categorical plot of kind-'swarm'

```python
sns.catplot(x='who',y='age',data=titanic,hue='sex',kind='swarm')
plt.show()
```

This is Categorical plot of kind-'violin'

```python
sns.catplot(x='who',y='class',data=titanic,hue='sex',kind='violin',palette='dark:red')
plt.show()
```

This is Categorical plot of kind-'box'

```python
sns.catplot(x='who',y='class',data=titanic,hue='sex',kind='box',palette='dark:red')
plt.show()
```

```python
sns.catplot(x='who',y='class',data=titanic,hue='sex',kind='bar',ec="white",palette='dark:red')
plt.show()
```

```python
sns.catplot(x='who',y='class',data=titanic,hue='sex',kind='point',palette='dark:red')
plt.show()
```
