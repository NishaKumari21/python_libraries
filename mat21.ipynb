Scatter Plot:A scatter plot is a graphical representation of data where each point on the graph represents a pair of values. It is commonly used to visualize the relationship between two variables. Scatter plots are useful for identifying patterns, trends, and outliers in the data.

Here are some key points about scatter plots:

1.
Each point on the scatter plot represents a pair of values (x, y).
2.
The position of each point on the graph is determined by the values of x and y.
3.
The scatter plot can be used to identify patterns, trends, and outliers in the data.
4.
The appearance of the scatter plot can be customized by changing colors, marker styles, and other visual properties.
5.
Scatter plots can be used to compare the relationship between multiple variables or to visualize the distribution of data.


```python
ls1=[25,35,68,24,61]
ls2=[96,45,12,36,45]
plt.scatter(x=ls1,y=ls2,marker="*")
plt.ylabel('List -2')
plt.xlabel("List-1")

plt.title("Scatter plot")
plt.show()
```

Histogram:It tells how our data is distributed.
A histogram plot is a graphical representation of data where the data is divided into bins and the frequency of each bin is represented by a bar. It is commonly used to visualize the distribution of data and identify patterns, trends, and outliers.

Here are some key points about histogram plots:

1.Each bar in the histogram represents a bin, which is a range of values.


2.The height of each bar represents the frequency or count of data points that fall within that bin.


3.The width of each bar can be adjusted to represent the size of the bins.


4.Histograms can be used to compare the distribution of data between different groups or categories.


5.Histograms can be customized by changing the color, style, and other visual properties.


6.Histograms can be used to identify patterns, trends, and outliers in the data.

Box plot:A box plot, also known as a box-and-whisker plot, is a graphical representation of data using a five-number summary (minimum, first quartile, median, third quartile, and maximum). It is commonly used to visualize the distribution of data and identify outliers.

Here are some key points about box plots:

1.The box represents the interquartile range (IQR), which is the range between the first quartile (25th percentile) and the third quartile (75th percentile).


2.The vertical lines extending from the box represent the minimum and maximum values, excluding outliers.


3.The median (second quartile) is represented by a horizontal line within the box.


4.Outliers are represented by individual points outside of the whiskers, which extend from the ends of the box.


5.Box plots can be used to compare the distribution of data between different groups or categories.


6.Box plots can be customized by changing the color, style, and other visual properties.

```python
plt.boxplot(age)
plt.show()
```
